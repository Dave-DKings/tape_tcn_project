# Glossary and Formula Sheet

Quick reference of key terms and equations.


## 1) Glossary

- **Simplex action**: non-negative weights summing to 1.
- **Turnover**: total absolute weight change per step.
- **DSR**: Differential Sharpe shaping term.
- **TAPE score**: weighted utility score at episode end.
- **Dual drawdown controller**: adaptive drawdown penalty mechanism.


## 2) Formula Sheet

Portfolio return:

$$
r_t = \frac{V_t - V_{t-1}}{V_{t-1}}.
$$

Turnover:

$$
\tau_t = \|w_t - w_{t-1}\|_1.
$$

Step reward:

$$
r_t^{\text{step}} = r_t^{\text{base}} + r_t^{\text{DSR}} + r_t^{\text{TO}} - r_t^{\text{DD}}.
$$

DSR:

$$
r_t^{\text{DSR}} = \lambda_{\text{DSR}}(\gamma S_t - S_{t-1}).
$$

Drawdown:

$$
DD_t = 1 - \frac{V_t}{V_t^{\text{peak}}}.
$$

Drawdown penalty:

$$
r_t^{\text{DD}} = c_{\text{dd}}\,\lambda_t\,\max(0,DD_t-DD^*).
$$

Terminal reward:

$$
r_T = \operatorname{clip}(\text{TAPE\_score}\cdot s_{\text{term}},-c,c).
$$


## 3) TCN Receptive Field Shortcut

For this code structure (two Conv1D layers per block):

$$
\text{RF} = 1 + 2(k-1)\sum_j d_j.
$$

Current default: $k=5$, $d=[2,4,8]$, so $\text{RF}=113$.
